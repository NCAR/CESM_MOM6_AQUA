```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import math

ds = xr.open_dataset('DATA_2300/23000101.ocean_annual.nc')

lat = ds['yq']
lon = ds['xh']
z = ds['zl']
vmo = ds['vmo']
vhGM = ds['vhGM']
```

```python
b = np.linspace(0, 30, 21, endpoint=True)
t = np.linspace(0, 30, 11, endpoint=True)
fig, ax = plt.subplots(figsize=(8,5))
plt.contourf(ds['xh'], ds['yh'], ds['tos'][-3,:,:], b, cmap='magma', extend='both')
cbar = plt.colorbar(ticks=t, boundaries=t, spacing='uniform')
cbar.ax.set_ylabel('[$^\circ$C]', fontsize=14)
plt.ylabel('Latitude', fontsize=14), plt.xlabel('Longitude', fontsize=14)
plt.ylim(-70,70)
plt.title('Coastline SST', fontsize=16)
plt.savefig('./FIGURES/coastlineSST_500yrs.png', bbox_inches='tight')
plt.show()
```

```python
inds = [None]*(81)
for lat in range(22,81):
        inds[lat] = np.where(np.isnan(vmo[0,0,lat,:]))
sums_small = []
sums_large = []
for lat in range(22,74):
#     print(lat)
    sfn_small = np.cumsum(np.nansum((vmo[:,:,lat,inds[lat][0][0]:inds[lat][0][-1]])/1025, axis=2), axis=1)
    sfn_large_1 = np.cumsum(np.nansum((vmo[:,:,lat,:inds[lat][0][0]])/1025, axis=2), axis=1) 
    sfn_large_2 = np.cumsum(np.nansum((vmo[:,:,lat,inds[lat][0][-1]:])/1025, axis=2), axis=1)
    sfn_large = sfn_large_1+sfn_large_2
    sums_small.append(sfn_small)
    sums_large.append(sfn_large)

sums_small = np.asarray(sums_small)
sums_large = np.asarray(sums_large)
print(sums_small.shape)    
print(sums_large.shape)
```

```python
sfn = vmo.sum('xh').cumsum('zl')/1025

sfn_so = np.cumsum(np.nansum((vmo[:,:,:23,:])/1025, axis=3), axis=1)
sfn_so_end = np.nanmean(sfn_so[-20:-10,:,:], axis=0)

v = np.linspace(-50, 50, 11, endpoint=True)
b = np.linspace(-50, 50, 11, endpoint=True)
c = np.linspace(-30, 30, 11, endpoint=True)
d = np.linspace(-10, 10, 11, endpoint=True)

plt.subplots(figsize=(8,5))
plt.contourf(ds['yq'], ds['zl'], np.nanmean(sfn[-10:-1,:,:], axis=0)/1e6, 
             v, cmap='coolwarm', extend='max')
plt.gca().invert_yaxis()
plt.xlim(-70, 70)
cbar = plt.colorbar(ticks=v, format = '%.0f')
cbar.ax.set_ylabel('[Sv]', fontsize=12)
plt.ylabel('Depth', fontsize=12), plt.xlabel('Latitude', fontsize=12)
plt.title('Coastline Mean Streamfunction', fontsize=16)
plt.savefig('./FIGURES/Coastline_Meansfn.pdf', bbox_inches='tight')
plt.show()

plt.subplots(figsize=(8,5))
plt.contourf(ds['yq'][22:74], z, (np.transpose(np.nanmean(sums_small[:,-10:,:],axis=1)))/1e6, 
             b, cmap='coolwarm', extend='max')
plt.gca().invert_yaxis()
cbar = plt.colorbar(ticks=b, format='%.0f')
cbar.ax.set_ylabel('[Sv]', fontsize=12)
plt.ylabel('Depth', fontsize=12), plt.xlabel('Latitude', fontsize=12)
plt.title('Coastline Small Basin Mean Streamfunction', fontsize=16)
plt.savefig('./FIGURES/Coastline_Meansfn_small.pdf', bbox_inches='tight')
plt.show()

plt.subplots(figsize=(8,5))
plt.contourf(ds['yq'][22:74], z, (np.transpose(np.nanmean(sums_large[:,-10:,:],axis=1)))/1e6, 
             c, cmap='coolwarm')
plt.gca().invert_yaxis()
cbar = plt.colorbar(ticks=c, format = '%.0f')
cbar.ax.set_ylabel('[Sv]', fontsize=12)
plt.ylabel('Depth', fontsize=12), plt.xlabel('Latitude', fontsize=12)
plt.title('Coastline Large Basin Mean Streamfunction', fontsize=16)
plt.savefig('./FIGURES/Coastline_Meansfn_large.pdf', bbox_inches='tight')
plt.show()

plt.subplots(figsize=(8,5))
plt.contourf(ds['yq'][:23], ds['zl'], sfn_so_end/(1e6), d, cmap='coolwarm')
plt.gca().invert_yaxis()
plt.xlim(-70, -35)
cbar = plt.colorbar(ticks=d, format = '%.0f')
cbar.ax.set_ylabel('[Sv]', fontsize=12)
plt.ylabel('Depth', fontsize=14), plt.xlabel('Latitude', fontsize=14)
plt.title('Coastline Southern Ocean Mean Streamfunction', fontsize=16)
plt.savefig('./FIGURES/Coastline_MeanSfn_so.pdf', bbox_inches='tight')
plt.show()
```

```python
inds = [None]*(81)
for lat in range(22,81):
        inds[lat] = np.where(np.isnan(vhGM[0,0,lat,:]))
sumsres_small = []
sumsres_large = []
for lat in range(22,74):
#     print(lat)
    sfnres_small = np.cumsum(np.nansum((vhGM[:,:,lat,inds[lat][0][0]:inds[lat][0][-1]])/1025, axis=2), axis=1)
    sfnres_large_1 = np.cumsum(np.nansum((vhGM[:,:,lat,:inds[lat][0][0]])/1025, axis=2), axis=1) 
    sfnres_large_2 = np.cumsum(np.nansum((vhGM[:,:,lat,inds[lat][0][-1]:])/1025, axis=2), axis=1)
    sfnres_large = sfnres_large_1+sfnres_large_2
    sumsres_small.append(sfnres_small)
    sumsres_large.append(sfnres_large)

sumsres_small = np.asarray(sumsres_small)
sumsres_large = np.asarray(sumsres_large)
print(sumsres_small.shape)    
print(sumsres_large.shape)
```

```python
sfnres = vhGM.sum('xh').cumsum('zl')/1025

sfnres_so = np.cumsum(np.nansum((vhGM[:,:,:23,:])/1025, axis=3), axis=1)
sfnres_so_end = np.nanmean(sfnres_so[-20:-10,:,:], axis=0)

a = np.linspace(-3, 3, 11, endpoint=True)
b = np.linspace(-2, 2, 11, endpoint=True)

plt.subplots(figsize=(8,5))
plt.contourf(ds['yq'], ds['zl'], np.nanmean(sfnres[-10:-1,:,:], axis=0)/1e6, 
             a, cmap='coolwarm')
plt.gca().invert_yaxis()
plt.xlim(-70, 70)
cbar = plt.colorbar(ticks=a, boundaries=a, spacing='uniform')
cbar.ax.set_ylabel('[Sv]', fontsize=12)
plt.ylabel('Depth', fontsize=12), plt.xlabel('Latitude', fontsize=12)
plt.title('Coastline Residual Streamfunction', fontsize=16)
plt.savefig('./FIGURES/Coastline_Ressfn.pdf', bbox_inches='tight')
plt.show()

plt.subplots(figsize=(8,5))
plt.contourf(ds['yq'][22:74], z, 
             (np.transpose(np.nanmean(sumsres_small[:,-10:,:],axis=1)))/1e6, 
             a, cmap='coolwarm')
plt.gca().invert_yaxis()
cbar = plt.colorbar(ticks=a, boundaries=a, spacing='uniform')
cbar.ax.set_ylabel('[Sv]', fontsize=12)
plt.ylabel('Depth', fontsize=12), plt.xlabel('Latitude', fontsize=12)
plt.title('Coastline Small Basin Residual Streamfunction', fontsize=16)
plt.savefig('./FIGURES/Coastline_Ressfn_small.pdf', bbox_inches='tight')
plt.show()

plt.subplots(figsize=(8,5))
plt.contourf(ds['yq'][22:74], z, 
             (np.transpose(np.nanmean(sumsres_large[:,-10:,:],axis=1)))/1e6, 
             a, cmap='coolwarm')
plt.gca().invert_yaxis()
cbar = plt.colorbar(ticks=a, boundaries=a, spacing='uniform')
cbar.ax.set_ylabel('[Sv]', fontsize=12)
plt.ylabel('Depth', fontsize=12), plt.xlabel('Latitude', fontsize=12)
plt.title('Coastline Large Basin Residual Streamfunction', fontsize=16)
plt.savefig('./FIGURES/Coastline_Ressfn_large.pdf', bbox_inches='tight')
plt.show()

plt.subplots(figsize=(8,5))
plt.contourf(ds['yq'][:23], ds['zl'], sfnres_so_end/(1e6), a, cmap='coolwarm')
plt.gca().invert_yaxis()
plt.xlim(-70, -35)
cbar = plt.colorbar(ticks=a, boundaries=a, spacing='uniform')
cbar.ax.set_ylabel('[Sv]', fontsize=12)
plt.ylabel('Depth', fontsize=14), plt.xlabel('Latitude', fontsize=14)
plt.title('Coastline Southern Ocean Residual Streamfunction', fontsize=16)
plt.savefig('./FIGURES/Coastline_Ressfnres_so.pdf', bbox_inches='tight')
plt.show()
```

```python
v = np.linspace(-30, 30, 11, endpoint=True)
b = np.linspace(-30, 30, 11, endpoint=True)
c = np.linspace(-30, 30, 11, endpoint=True)
d = np.linspace(-20, 20, 11, endpoint=True)

plt.subplots(figsize=(8,5))
plt.contourf(ds['yq'], ds['zl'], 
             (np.nanmean(sfn[-10:-1,:,:], axis=0) + np.nanmean(sfnres[-10:-1,:,:], axis=0))/1e6, 
             v, cmap='coolwarm', extend='both')
plt.gca().invert_yaxis()
plt.xlim(-70, 70)
cbar = plt.colorbar(ticks=v, format = '%.0f')
cbar.ax.set_ylabel('[Sv]', fontsize=12)
plt.ylabel('Depth', fontsize=12), plt.xlabel('Latitude', fontsize=12)
plt.title('Coastline Residual Mean Streamfunction', fontsize=16)
plt.savefig('./FIGURES/Coastline_ResMeansfn.pdf', bbox_inches='tight')
plt.show()

plt.subplots(figsize=(8,5))
plt.contourf(ds['yq'][22:74], z, 
             ((np.transpose(np.nanmean(sums_small[:,-10:,:],axis=1)))+(np.transpose(np.nanmean(sumsres_small[:,-10:,:],axis=1))))/1e6, 
             b, cmap='coolwarm', extend='both')
plt.gca().invert_yaxis()
cbar = plt.colorbar(ticks=b, format='%.0f')
cbar.ax.set_ylabel('[Sv]', fontsize=12)
plt.ylabel('Depth', fontsize=12), plt.xlabel('Latitude', fontsize=12)
plt.title('Small Basin Residual Mean Streamfunction', fontsize=16)
plt.savefig('./FIGURES/Coastline_ResMeansfn_small.pdf', bbox_inches='tight')
plt.show()

plt.subplots(figsize=(8,5))
plt.contourf(ds['yq'][22:74], z, 
             ((np.transpose(np.nanmean(sums_large[:,-10:,:],axis=1)))+(np.transpose(np.nanmean(sumsres_large[:,-10:,:],axis=1))))/1e6, 
             c, cmap='coolwarm', extend='both')
plt.gca().invert_yaxis()
cbar = plt.colorbar(ticks=c, format = '%.0f')
cbar.ax.set_ylabel('[Sv]', fontsize=12)
plt.ylabel('Depth', fontsize=12), plt.xlabel('Latitude', fontsize=12)
plt.title('Large Basin Residual Mean Streamfunction', fontsize=16)
plt.savefig('./FIGURES/Coastline_ResMeansfn_large.pdf', bbox_inches='tight')
plt.show()

plt.subplots(figsize=(8,5))
plt.contourf(ds['yq'][:23], ds['zl'], 
             (sfn_so_end+sfnres_so_end)/(1e6), 
             d, cmap='coolwarm', extend='both')
plt.gca().invert_yaxis()
plt.xlim(-70, -35)
cbar = plt.colorbar(ticks=d, format = '%.0f')
cbar.ax.set_ylabel('[Sv]', fontsize=12)
plt.ylabel('Depth', fontsize=14), plt.xlabel('Latitude', fontsize=14)
plt.title('Southern Ocean Residual Mean Streamfunction', fontsize=16)
plt.savefig('./FIGURES/Coastline_ResMeanSfn_so.pdf', bbox_inches='tight')
plt.show()
```
