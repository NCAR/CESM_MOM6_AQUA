```python
import matplotlib.pyplot as plt
import xarray as xr
import numpy as np

f0 = xr.open_dataset('DATA_2400_new/24000101.ocean_annual.nc', decode_times=False)
f1 = xr.open_dataset('DATA_2300/23000101.ocean_annual.nc', decode_times=False)
f2 = xr.open_dataset('DATA_2200/22000101.ocean_annual.nc', decode_times=False)
f3 = xr.open_dataset('DATA_2100/21000101.ocean_annual.nc', decode_times=False)
f4 = xr.open_dataset('DATA_2000/20000101.ocean_annual.nc', decode_times=False)
f5 = xr.open_dataset('DATA_1900/19000101.ocean_annual.nc', decode_times=False)
```

```python
sst = np.concatenate((f5['tos'][:], f4['tos'][:], f3['tos'][:], f2['tos'][:], f1['tos'][:], f0['tos'][:]))
tmp = np.nanmean(sst, axis=1)
mean_sst = np.nanmean(tmp, axis=1)

plt.plot(mean_sst)
plt.title('Global Mean Sea Surface Temperature')
plt.ylabel('[$^o$C]'), plt.xlabel('Year')
plt.show()

print('Sea Surface Temperature change per 100 years')
print('0-100:', mean_sst[99]-mean_sst[0])
print('100-200:', mean_sst[199]-mean_sst[100])
print('200-300:', mean_sst[299]-mean_sst[200])
print('300-400:', mean_sst[399]-mean_sst[300])
print('400-500:', mean_sst[499]-mean_sst[400])
print('500-540:', mean_sst[539]-mean_sst[500])
```

```python
temp = np.concatenate((f5['thetao'][:], f4['thetao'][:], f3['thetao'][:], f2['thetao'][:], f1['thetao'][:], f0['thetao'][:]))
tmp = np.nanmean(temp[:,25,:,:], axis=1)
mean_dtemp = np.nanmean(tmp, axis=1)

plt.plot(mean_dtemp)
plt.title('2000m Global Mean Temperature')
plt.ylabel('[$^o$C]'), plt.xlabel('Year')
plt.show()

print('2000m Temperature change per 100 years')
print('0-100:', mean_sst[99]-mean_sst[0])
print('100-200:', mean_sst[199]-mean_sst[100])
print('200-300:', mean_sst[299]-mean_sst[200])
print('300-400:', mean_sst[399]-mean_sst[300])
print('400-500:', mean_sst[499]-mean_sst[400])
print('500-540:', mean_sst[539]-mean_sst[500])
```
